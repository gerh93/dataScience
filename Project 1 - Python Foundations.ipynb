### Uni-variate Graph Analysis

```python
# Analysis for Product
# Function to annotate the percentage for data in x-axis
def var_perc (plot, feature):
    total = len(feature)
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height()/total)
        x=p.get_x() + p.get_width()/2-0.05
        y=p.get_y() + p.get_height()
        ax.annotate(percentage, (x,y), size =15)

plt.figure(figsize=(7,7))
ax = sns.countplot(copy_df['Product'], palette="PuBuGn_r")
var_perc(ax, copy_df['Product'])

sns.displot(copy_df['Product'], kde=True);
```

```python
# Analysis for Marital Satus 
plt.figure(figsize=(7,7))
ax = sns.countplot(copy_df['MaritalStatus'], palette="PuBuGn")
var_perc(ax, copy_df['MaritalStatus'])

sns.displot(copy_df['MaritalStatus'], kde=True);
```

```python
# Analysis for Education
plt.figure(figsize=(7,7))
ax = sns.countplot(copy_df['Education'], palette="PuBuGn")
var_perc(ax, copy_df['Education'])

sns.displot(copy_df['Education'], kde=True);
```

```python
# Analysis for Age
plt.figure(figsize=(15,7))
ax = sns.countplot(copy_df['Age'], palette = "mako_r")
var_perc(ax, copy_df['Age'])

sns.displot(copy_df['Age'], kde=True);
```

```python
# Analysis for Gender
plt.figure(figsize=(15,7))
ax = sns.countplot(copy_df['Gender'], palette = "mako_r")
var_perc(ax, copy_df['Gender'])

sns.displot(copy_df['Gender'], kde=True);
```

```python
# Analysis for Usage
plt.figure(figsize=(15,7))
ax = sns.countplot(copy_df['Usage'], palette = "mako_r")
var_perc(ax, copy_df['Usage'])

sns.displot(copy_df['Usage'], kde=True);
```

## Conclusion
I analyzed the data of 3 different products offered by Cardio Good Fitness (the "Firm"), utilizing the customer data set given to me by the Firm. Through my research and observation, I have come to the conclusions that the average customer profile is a partnered male between 24-34 years old, with 16 years of education and an annual income of 55,000.

The analysis conclusions are listed below:

1. The highest selling product is TM195.
2. The majority of customer that purchase product TM195 fall between the ages of 24-34.
3. The majority of customers that have had 13-17 years of education purchase product TM195 and TM498. 
4. Customers with higher education and higher income gravitate to product TM798. 
5. The majority of single female customers that stated they utilized their treadmill 2-4 times a week purchased Product TM195.
6. The majority of partnered male customers that stated they utilized their treadmill 2-4 times a week purchased product TM498 
7. The majority of single male customers that stated they utilized their treadmill 4-6 times a week purchased Product TM798. 
8. Customers between the age of 40-50 years old purchased less treadmills than customers within the age group of 20-38. 
9. Customers between 40-50 years old purchased treadmills once they reached an income of at least ~50,000.   

## Recommendations to Cardio Good Fitness

1. Since product TM195 is the highest selling product, apply a significant portion of the Firm's marketing budget on individuals that are part of Gen Y (24-40 year old), with at least an associates degree, starting a career with an entry level position (income of 30,000 - 60,000).  


2. Discontinue the production of product TM498. This product resonates with the same audience as product TM195 with lower sales. The phasing out of product TM498 will streamline the Firm's products offering, allowing management to efficiently control its resources and outputs.


3. Product TM798 resonates with a similar age demographic as TM195. However, these customers have a higher income (Income of 50,000 to 100,000) and are more educated than purchasers of product TM195. Therefore, target higher income individuals within Gen Y (24-40 year old) population and make sure to upsale and educate individuals on the different features between product TM195 and TM798. 


4. The product offerings of the Firm resonate heavily with individuals between the ages of 22-35, with little interest from the older individuals (40+ years old). The Firm would benefit from creating a new product that focuses on the older demographic (40-50 year old) to capture a new market.
